import libraries

```python
import pandas as pd
import numpy as np
```

load and summarize dataset

```python
df = pd.read_csv('dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

segregate dataset into x and y

```python
x = df.iloc[:, :-1].values
x
```

```python
y = df.iloc[:, -1].values
y
```

spliiting dataset to tarin and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

feature scaling (standardization - zscore & normalization - min_max_normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

model training

```python
from sklearn.linear_model import LogisticRegression

# loading the algorithm
model = LogisticRegression() 

# train the model
model.fit(x_train, y_train)
```

```python
# using help() we can check that the attributes of the model

help(LogisticRegression)
```

validation process

```python
y_pred =  model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy of the model : {:.4f}%".format(accuracy_score(y_test, y_pred)*100))
```

prediction

```python
age = int(input("Enter new customer's age: "))
salary = int(input("Enter new customer's salary: "))
newCust = [[age, salary]]
res = model.predict(scaler.transform(newCust))
print(res)
if res == 1:
    print("Customer will buy.")
else:
    print("Customer won't buy.")
```
