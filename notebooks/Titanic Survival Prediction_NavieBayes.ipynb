import libraries

```python
import pandas as pd
import numpy as np
```

load the dataset and summerization

```python
dataset = pd.read_csv('titanic.csv')
```

```python
print(dataset.shape)
dataset.head()
```

mapping / converting categorical data to binary values

```python
# There are 2 values in the sex column such as male and female. we can map them into binary value(0 and 1).

data_set = set(dataset['Sex'])
dataset['Sex'] = dataset['Sex'].map({'female':0, 'male': 1}).astype(int)
dataset.head()
```

segregate dataset into x(input) and y(output) 

```python
x = dataset.drop('Survived', axis=1)
x
```

```python
y = dataset.Survived
y
```

find the NA values in the dataset and fill / remove from them our feature 

```python
# checking whether there are missing values of any columns

x.columns[x.isna().any()]
```

* In the above output shows that age column has missing values

```python
# we fill them with mean of the Age colunm

x.Age = x.Age.fillna(x.Age.mean())
```

```python
x.columns[x.isna().any()]
```

splitting dataset into train and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

training the model - Naive Bayes (Gaussian Naive Bayes)

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)
```

```python
# check the more details about Gaussian naive bayes

help(GaussianNB)
```

prediction for the all test data

```python
y_pred = model.predict(x_test)
print(np.column_stack((y_pred, y_test)))
```

accuracy of the model

```python
from sklearn.metrics import accuracy_score

print("Accuracy of the model: {0}%".format(accuracy_score(y_test, y_pred)*100))
```

predicting whether person survived or not

```python
pclassNo = int(input("Enter Person's Pclass number: "))
gender = int(input("Enter Person's Gender, female-0 & male-1(0 or 1): "))
age = int(input("Enter person's Age: "))
fare = float(input("Enter Person's Fare: "))
person = [[pclassNo, gender, age, fare]]
res = model.predict(person)
print(res)

if res == 1:
    print("Person might be survived.")
else:
    print("Person might not be survived.")
```
